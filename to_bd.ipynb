```python
# Connexion à la base de données SQL Server
conn = pyodbc.connect("Driver={SQL Server};Server=LAPTOP-IGO4LBSV\SQLEXPRESS;Database=github_db;Trusted_Connection=yes;")
cursor = conn.cursor()

cursor.execute("SELECT @@version;")
row = cursor.fetchone()
while row: 
    print(row[0])
    row = cursor.fetchone()
```

```python
# Insertion des données dans la table "Owner"
owner_table = data[["owner"]].drop_duplicates().reset_index(drop=True)

owner_table.insert(0, "id_owner", range(1, len(owner_table) + 1))
#insert_owner_data("Owner", owner_table[["id_owner", "owner"]])
display(owner_table)
```

```python
# Insertion des données de la table "Owner" dans la base de donnees

def insert_owner_data(cursor, owner_data):
    for index, row in owner_data.iterrows():
        cursor.execute("INSERT INTO Owners (id_owner, name_o) VALUES (?, ?)", row["id_owner"], row["owner"])
    conn.commit()

    # Appel de la fonction pour insérer les données dans la table "Owner"
insert_owner_data(cursor, owner_table)

# Fermeture de la connexion à la base de données
conn.close()
```

```python
# Création de la table language_table (id_langage sera généré automatiquement)
language_table = data[["language"]].drop_duplicates().reset_index(drop=True)
language_table.insert(0, "id_langage", range(1, len(language_table) + 1))
display(language_table)
```

```python
#la relation entre langage et repo
for index,L in language_table.iterrows():
    data['language'].replace(L['language'],L['id_langage'],inplace=True)
```

```python
# la relation entre owner et repo
for index,O in owner_table.iterrows():
    data['owner'].replace(O['owner'],O['id_owner'],inplace=True)
```

```python
# Création de la table repo_table (id_repo sera généré automatiquement)
repo_table = data[["name", "url", "description", "stars", "created_at", "forks", "watchers", "open_issues", "contributors_count", "language", "owner"]].reset_index(drop=True)
repo_table.insert(0, "id_repo", range(1, len(repo_table) + 1))
```

```python
# Insertion des données dans la table "Repo"
def insert_langage_data(cursor, repo_data):
    for index, row in repo_data.iterrows():
        cursor.execute("INSERT INTO Repo (id_repo, name_repo, url, description, stars, created_at, forks, watchers, open_issues, contributors_count, id_langage, id_owner) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)", row["id_repo"], row["name"], row["url"], row["description"], row["stars"], row["created_at"], row["forks"], row["watchers"], row["open_issues"], row["contributors_count"], row["language"], row["owner"])
    conn.commit()

# Appel de la fonction pour insérer les données dans la table "Repo"
insert_langage_data(cursor, repo_table)

# Fermeture de la connexion à la base de données
conn.close()

```

```python
# Insertion des données dans la table "Langage"
def insert_langage_data(cursor, langage_data):
    for index, row in langage_data.iterrows():
        cursor.execute("INSERT INTO Langage (id_langage, langage) VALUES (?, ?)", row["id_langage"], row["language"])
    conn.commit()

# Appel de la fonction pour insérer les données dans la table "Langage"
insert_langage_data(cursor, language_table)

# Fermeture de la connexion à la base de données
conn.close()
```

```python
# Création de la table language_table (id_langage sera généré automatiquement)
Contributors_table = data[["contributors"]].drop_duplicates().reset_index(drop=True)
Contributors_table.insert(0, "id_contributors", range(1, len(Contributors_table) + 1))
display(Contributors_table)
```

```python
data1 = data[['url', 'contributors', 'contributions_count']]
data1 = data1.sort_values(by="url")
repos1 = repo_table[['id_repo', 'url']]
repos1 = repos1.sort_values(by="url")
repos1 = repos1.drop_duplicates()
```

```python
data1['id_repo'] = repos1['id_repo']
data1 = data1.explode('contributors')
```

```python
contributions = data[['url', 'contributors']]
contributions['contributors'].astype(str)
# split the contributors column
contributions['contributors'] = contributions['contributors'].astype(str).str.split(',')
# explode the contributors column
contributions = contributions.explode('contributors')
# remove the spaces
contributions['contributors'] = contributions['contributors'].str.strip()
# remove the quotes
contributions['contributors'] = contributions['contributors'].str.replace("'", "")

contributions1 = data[['url', 'contributions_count']]
# split the contributions_count column
contributions1['contributions_count'] = contributions1['contributions_count'].astype(str).str.split(',')
# explode the contributions_count column
contributions1 = contributions1.explode('contributions_count')
# remove the spaces
contributions1['contributions_count'] = contributions1['contributions_count'].str.strip()
# explode the contributions_count column
contributions1 = contributions1.explode('contributions_count')

contributions["contributions_count"]  = contributions1["contributions_count"]

```

```python
contributions['contributors'] = contributions['contributors'].str.replace("'", "")
contributions['contributors'] = contributions['contributors'].str.replace("\"", "")
contributions['contributors'] = contributions['contributors'].str.replace("[", "")
contributions['contributors'] = contributions['contributors'].str.replace("]", "")
contributions['contributors'] = contributions['contributors'].str.strip()

contributions['contributions_count'] = contributions['contributions_count'].str.replace("[", "")
contributions['contributions_count'] = contributions['contributions_count'].str.replace("]", "")
contributions['contributions_count'] = contributions['contributions_count'].str.strip()
```

```python
# Insertion des données dans la table "Owner"
contributors_table = contributions[["contributors"]].drop_duplicates().reset_index(drop=True)

contributors_table.insert(0, "id_Contributors", range(1, len(contributors_table) + 1))
#insert_owner_data("Owner", owner_table[["id_owner", "owner"]])
display(contributors_table)
```

```python
# Insertion des données dans la table "Contributors" 

def insert_contributors_data(cursor, contributors_data):
    for index, row in contributors_data.iterrows():
        cursor.execute("INSERT INTO Contributors (id_Contributors, name_Contributors) VALUES (?, ?)", row["id_Contributors"], row["contributors"])
    conn.commit()

    # Appel de la fonction pour insérer les données dans la table "Contributors"
insert_contributors_data(cursor, contributors_table)

# Fermeture de la connexion à la base de données
conn.close()
```

```python
merged_df = contributions.merge(repo_table, on='url', how='left')
merged_df1 = merged_df.merge(contributors_table, on='contributors', how='left')
merged_df1 = merged_df1[['id_repo', 'id_Contributors', 'contributions_count']]
merged_df1
```

```python
# Insertion des données dans la table "Contributors_repo" 

def insert_Contributors_repo_data(cursor, contributors_repo_data):
    for index, row in contributors_repo_data.iterrows():
        cursor.execute("INSERT INTO Contributors_repo (id_Contributor, id_Repo, contribution_count) VALUES (?, ?, ?)", row["id_Contributors"], row["id_repo"], row["contributions_count"])
    conn.commit()

    # Appel de la fonction pour insérer les données dans la table "Contributors"
insert_Contributors_repo_data(cursor, merged_df1)

# Fermeture de la connexion à la base de données
conn.close()
```
