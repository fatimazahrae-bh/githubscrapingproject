```python
headers = {'Authorization': 'g'}

# les repositories qui on 2 ou plus stars
#query = 'stars:>1'
date_debut = datetime(2023, 1, 1)
date_fin = datetime.now()
current_date = date_debut

while current_date <= date_fin:
     # Convert date to string format
    date_str = current_date.strftime('%Y-%m-%d') 
    print("Trending repositories for:", date_str)
     # ==========================
    url = f'https://github.com/trending?{date_str}'    
    
    response = requests.get(url, headers=headers)
    result = response.json()['items']

    # Count the number of repositories using each programming language
    languages = []
    for repo in result:
        url = repo['languages_url']
        response = requests.get(url, headers=headers)
        languages.extend(response.json().keys())
        language_counts = Counter(languages)

    # Print the top 10 programming languages by repository count
    for language, count in language_counts.most_common(10):
        print(language, count)
    current_date += timedelta(days=1)
```
